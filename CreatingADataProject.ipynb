```python
# CLEANING COLUMNS: Eliminate String characters from numeric data and changing them to floats

for i in range(0, len(data)):
    if(' day ago' in data['Uploaded'][i] or ' days ago' in data['Uploaded'][i]):
        s = (data['Uploaded'][i].split( 'day')[0])
        data['Uploaded'][i] = float(s) * (1/30)
    elif(' week ago' in data['Uploaded'][i] or ' weeks ago' in data['Uploaded'][i]):
        s = (data['Uploaded'][i].split( ' week')[0])
        data['Uploaded'][i] = float(s) * .25 
    elif(' month ago' in data['Uploaded'][i] or ' months ago' in data['Uploaded'][i]):
        s = (data['Uploaded'][i].split( ' month')[0])
        data['Uploaded'][i] = float(s)
    elif(' year ago' in data['Uploaded'][i] or ' years ago' in data['Uploaded'][i]):
        s = (data['Uploaded'][i].split(' year')[0]) 
        data['Uploaded'][i] = float(s) * 12
            
for i in range(0, len(data2)):
    if(' day ago' in data2['Uploaded'][i] or ' days ago' in data2['Uploaded'][i]):
        s = (data2['Uploaded'][i].split( 'day')[0])
        data2['Uploaded'][i] = float(s) * (1/30)
    elif(' week ago' in data2['Uploaded'][i] or ' weeks ago' in data2['Uploaded'][i]):
        s = (data2['Uploaded'][i].split( ' week')[0])
        data2['Uploaded'][i] = float(s) * .25 
    elif(' month ago' in data2['Uploaded'][i] or ' months ago' in data2['Uploaded'][i]):
        s = (data2['Uploaded'][i].split( ' month')[0])
        data2['Uploaded'][i] = float(s)
    elif(' year ago' in data2['Uploaded'][i] or ' years ago' in data2['Uploaded'][i]):
        s = (data2['Uploaded'][i].split(' year')[0]) 
        data2['Uploaded'][i] = float(s) * 12

data = data[data['Views'].str.contains(',')]
data2 = data2[data2['Views'].str.contains(',')]

data['Views'] = data['Views'].str.replace(',','')
data2['Views'] = data2['Views'].str.replace(',','')
data['Views'] = data['Views'].str.replace('Not included', '')
data2['Views'] = data2['Views'].str.replace('Not included', '')
data['Views'] = data['Views'].astype(float)
data2['Views'] = data2['Views'].astype(float)
        
```

```python
# Create a Minutes and a Seconds column
data['Minutes'] = pd.Series((length.split(':')[0] for length in data['Duration']), index = data.index).astype(float)
data['Seconds'] = pd.Series((length.split(':')[1] for length in data['Duration']), index = data.index).astype(float)
# Create the column "New Duration" that includes the duration of each video in minutes.
data['New Duration'] = data.Minutes + data.Seconds/60

# Display the New Duration column
data['New Duration']
```

###### 3. Merge dataframes from separate json files as appropriate
- Find the intersection of two (or more) sets
- Compare the intersection with the newer set to find 'New Products'  
    - When found, add the starting date
- Compare the intersection with the older set to find 'Closed Products'  
    - When found, add the closing date
    - Compare closing date with starting date to find days on market      

```python
# Create the inner merge of the two dataframes (the videos in common from data and data2)
dataInnerMerge = pd.merge(data, data2, \
                          on=['Title', 'Channel', 'Duration'],\
                          how='inner')

# Display the videos in common from data and data2
dataInnerMerge
```

```python
# Determine the set of new videos (the videos that were in data2, but not in data)
data2['key1'] = 1
dataInnerMerge['key2'] = 1
newVideos = pd.merge(data2, dataInnerMerge, \
                      on=['Title', 'Channel', 'Duration'], \
                      how = 'left')

# Create the associated dataframe
newVideos = newVideos[newVideos.key2 != newVideos.key1]

# Display the new videos (the ones that were in data2, but not in data)
newVideos
```

```python
# Determine the set of closed videos (the videos in data, but not in data2)
data['key1'] = 1
closedVideos = pd.merge(data, dataInnerMerge, \
                      on=['Title', 'Channel', 'Duration'], \
                      how = 'left')

# Create the associated dataframe
closedVideos = closedVideos[closedVideos.key2 != closedVideos.key1]

# Display the closed videos (the ones that were in data, but not in data2
closedVideos
```

```python
# Find the average views per day per video in dataInnerMerge and add it as a new column
dataInnerMerge['Average View Increase Per Day'] = (dataInnerMerge.Views_y - dataInnerMerge.Views_x)/7
dataInnerMerge
```

```python
# Add a column which counts the number of entries in the list an artist has
dataInnerMerge['Artist Entries On List'] = 0
for i in range(0, len(dataInnerMerge)):
    channel = dataInnerMerge['Channel'][i]
    for j in range(0, len(dataInnerMerge)):
        if channel == dataInnerMerge['Channel'][j]:
            dataInnerMerge['Artist Entries On List'][i] = dataInnerMerge['Artist Entries On List'][i] + 1
dataInnerMerge
```

```python
# Create a Scatter Diagram of Views and Uploaded
fig = plt.figure()
ax = fig.add_subplot(111)
x = dataInnerMerge.Uploaded_y
y = dataInnerMerge.Views_y
ax.scatter(x,y,color='blue')
ax.set_xlim([0,12])
ax.set_ylim([0,1000000000])

plt.title("Views vs. How Many Months Ago Uploaded")
plt.xlabel("Uploaded x Months Ago")
plt.ylabel("Views in Billions")
plt.show()
```

```python
# Create a Scatter Diagram of the weekly change in Views and how many months ago the video was Uploaded
dataInnerMerge['View Difference'] = dataInnerMerge.Views_y - dataInnerMerge.Views_x

fig = plt.figure()
ax = fig.add_subplot(111)
x = dataInnerMerge.Uploaded_y
y = dataInnerMerge['View Difference']
ax.scatter(x,y,color='blue')
ax.set_xlim([0,12])
ax.set_ylim([0,40000000])

plt.title("Weekly Change in Views vs. How Many Months Ago Uploaded")
plt.xlabel("Uploaded x Months Ago")
plt.ylabel("Weekly Change in Views in Tens of Millions")
plt.show()
```

```python
# Create a Scatter Diagram of Views and Uploaded
fig = plt.figure()
ax = fig.add_subplot(111)
x = dataInnerMerge.Uploaded_y
y = dataInnerMerge['Average View Increase Per Day']
ax.scatter(x,y,color='blue')
ax.set_xlim([0,12])
ax.set_ylim([0,10000000])

plt.title("Average Daily Views vs. How Many Months Ago Uploaded")
plt.xlabel("Uploaded x Months Ago")
plt.ylabel("Average Views Per Day in Tens of Millions")
plt.show()
```

```python
# Create a histogram of how many months ago the top music videos from last year were uploaded
plt.hist(dataInnerMerge.Uploaded_y)
plt.title("How Many Months Ago Were Top Music Videos From the Last Year Uploaded")
plt.xlabel("Uploaded x Months Ago")
plt.ylabel("Frequency")
plt.show()
```

```python
# Show in a pie chart the percentage of music videos that come from a Vevo channel
VevoVideos = 0 
for i in range(0, len(dataInnerMerge)):
    if ('VEVO' in dataInnerMerge['Channel'][i] or 'Vevo' in dataInnerMerge['Channel'][i]):
        VevoVideos = VevoVideos + 1
NotVevo = len(dataInnerMerge) - VevoVideos

labels = 'VEVO', 'Not VEVO'
sizes = [VevoVideos, NotVevo]
colors = ['gold', 'yellowgreen']
explode = (0.1, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
# Show in a pie chart the percentage of views from the top music videos that
# come from Vevo channels versus non-Vevo channels
VevoVideos = 0 
for i in range(0, len(dataInnerMerge)):
    if ('VEVO' in dataInnerMerge['Channel'][i] or 'Vevo' in dataInnerMerge['Channel'][i]):
        VevoVideos = VevoVideos + dataInnerMerge['Views_y'][i]

AllVideos = 0
for i in range(0, len(dataInnerMerge)):
    AllVideos = AllVideos + dataInnerMerge['Views_y'][i]
    
NotVevo = AllVideos - VevoVideos

labels = 'VEVO', 'Not VEVO'
sizes = [VevoVideos, NotVevo]
colors = ['gold', 'yellowgreen']
explode = (0.1, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
